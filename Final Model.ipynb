```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics
from sklearn.linear_model import Lasso, Ridge
from sklearn.model_selection import GridSearchCV, train_test_split
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
```

```python
df=pd.read_csv('clean_data')
df.drop('Unnamed: 0',axis=1,inplace=True)
```

```python
df['constructionTime'] = df['constructionTime'].astype('object')
```

```python
df.info()
```

```python
df.drop(['url','id','Lat','Lng','Cid','tradeTime','DOM','followers','communityAverage','fiveYearsProperty'],inplace=True,axis=1)
```

```python
df.info()
```

```python
from sklearn.preprocessing import RobustScaler
```

```python
robust = RobustScaler()
```

```python
## Setting features and target
features = df.drop('totalPrice',axis=1)
features = pd.get_dummies(features,drop_first=True)
## Lanjutan standarisasi nilai
robust.fit(features)
robust_data = robust.transform(features)
features = pd.DataFrame(robust_data,columns=features.columns)
target = df['totalPrice']
```

```python
## Train test split
x_train, x_test, y_train, y_test = train_test_split(features, target, test_size=0.1, random_state=101)
```

```python
rf = RandomForestRegressor(random_state=101,
                           max_depth =  None, 
                           min_samples_leaf =  5, 
                           min_samples_split = 6,
                           n_estimators =  120
                           )
```

```python
rf.fit(x_train,y_train)
```

```python
# Prediction with data train
predictions_train = rf.predict(x_train)
print('RMSE : ', np.sqrt(metrics.mean_squared_error(y_train,predictions_train)))
print('R2 Score : ', r2_score(y_train,predictions_train))
f, axes = plt.subplots(1,2,figsize=(20,10))
sns.regplot(x = y_train, y = predictions_train,ax = axes[0])
sns.distplot((y_train-predictions_train),ax = axes[1])
```

```python
# Prediction with data test
predictions_test = rf.predict(x_test)
print('RMSE : ', np.sqrt(metrics.mean_squared_error(y_test,predictions_test)))
print('R2 Score : ', r2_score(y_test,predictions_test))
f, axes = plt.subplots(1,2,figsize=(20,10))
sns.regplot(x = y_test, y = predictions_test,ax = axes[0])
sns.distplot((y_test-predictions_test),ax = axes[1])
```
